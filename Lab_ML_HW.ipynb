Набор данных представляет из себя параметры объектов, обнаруженных наземным черенковским телескопом MAGIC. Он обнаруживает потоки частиц, выпущенных гамма-лучами, используя эффект Черенкова-Вавилова, то есть излучение, вызываемое в прозрачной среде заряженной частицей, которая движется со скоростью, превышающей фазовую скорость распространения света в этой среде.
 
 # Title of Database: MAGIC gamma telescope data 2004

Attribute information:

1.  fLength:   continuous  # major axis of ellipse [mm] 
                             = Большая ось эллипса (в мм)
2.  fWidth:    continuous  # minor axis of ellipse [mm] 
                             = Малая ось эллипса (в мм)
3.  fSize:     continuous  # 10-log of sum of content of all pixels [in #phot] 
                             = Десятичный логарифм суммы значений всех пикселей (в фотах)
4.  fConc:     continuous  # ratio of sum of two highest pixels over fSize  [ratio]
                             = Соотнношение суммы двух наивысших значений пикселей с fSize (соотношение)
5.  fConc1:    continuous  # ratio of highest pixel over fSize  [ratio]
                             = Соотнношение наивысшего значения пикселея с fSize (соотношение)
6.  fAsym:     continuous  # distance from highest pixel to center, projected onto major axis [mm]
                             = Расстояние от пикселя с наивысшем значением до центра, спроецированное на большую ось (в мм)
7.  fM3Long:   continuous  # 3rd root of third moment along major axis  [mm]
                             = Третий корень третьего момента вдоль большой оси эллипса (в мм)
8.  fM3Trans:  continuous  # 3rd root of third moment along minor axis  [mm]
                             = Третий корень третьего момента вдоль малой оси эллипса (в мм)
9.  fAlpha:    continuous  # angle of major axis with vector to origin [deg]
                             = Угол большой оси эллипса с вектором до начала (в градусах)
10.  fDist:    continuous  # distance from origin to center of ellipse [mm]
                             = Расстояние от начала до центра эллипса (в мм)
11.  class:    g,h         # gamma (signal), hadron (background)
                             = Класс объекта - сигнал или фоновый

# Метод к-ближайших соседей

Выбор параметр равный 10:

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
#knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
knn
```

Классификация и подсчёт ошибки:

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

Другой способ подсчёта ошибки:

```python
err_test = 1 - knn.score(X_test, y_test)
```

```python
err_test
```

## Подбор параметра
Минимизируем ошибку, подобрав значение параметра:

```python
# Долго!
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

Вычислим значение ошибки с найденным параметром:

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

Тренировочная ошибка уменьшилась, тестовая осталась такой же. Ошибка определения принадлежности объекта к классу (сигнал или фоновый) составляет 0.16579039607430773.
